```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
def plot_board(board):
    plt.imshow(board, aspect='auto')
    plt.colorbar()
    plt.show()
```

```python
def plot_df(df):
    df.plot(x='episode', y='reward', style='o')
    plt.show()
    df['reward'].hist()
    plt.show()
```

```python
def summarize_episodes(path, frequency=100):
    files = os.listdir(path)
    i = 0
    df = []
    for f in files:
        if f[-3:] == 'npy':
            record = {'filename': f}
            fs = f[:-4].split('_')
            i += 1
            board = np.load(os.path.join(path, f))
            record['episode'] = int(fs[1])
            record['reward'] = float(fs[3])
            df.append(record)
            if i % frequency == 0:
                print('file:', f)
                plot_board(board)
    df = pd.DataFrame(df)
    print(df)
    plot_df(df)
    return df
```

```python
df = summarize_episodes('./output/', frequency=500)
```
